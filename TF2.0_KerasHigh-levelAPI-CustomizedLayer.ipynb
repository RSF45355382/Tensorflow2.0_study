```python
import tensorflow as tf 
import tensorflow.keras as keras
```

```python
# 构建一个神经网络
network = keras.Sequential([
    keras.layers.Dense(256,activation='relu'),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(10),
])

# 这一句才会初始化所有网络中的权值以及偏置参数
network.build(input_shape = (None,28,28))
network.summary()
```

```python
# Sequential还可以方便的管理他内部的参数

# 这一句反馈其中所有的训练可变参数
network.trainable_variables
```

```python
network.call()
# 这里是进行前向传播的接口, 他会自动将input从头转到尾, 并反馈输出
# 这一句和network(input)是一样的
# network(input) 这一句实质上是调用了这个对象的__call__()方法, 而model的__call__()方法中调用了call()方法
```

## Sequential其实就是keras.Model的一个子类, 一个模型要想使用compile/predict/fit/evaluate这些高级接口的话, 必须要继承自keras.Model这个父类
