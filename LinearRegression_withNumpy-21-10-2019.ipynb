```python
# 使用numpy实现线性回归
import numpy as np 
import pandas as pd 
import random
# 学习率
lr = 0.000001

data = pd.read_excel('LinearRegression_data.xlsx',sheet_name='Sheet1')
# 获取numpy.array数据
data = data.values
print(data[:10,:])
num_sample = data.shape[0]
print(num_sample)
x = data[:,0]
# 制作输入数据, 加一列全部元素为1的数据
input_x = np.hstack((x.reshape([-1,1]),np.ones([data.shape[0],1])))
# 
y = data[:,1].reshape([-1,1])
print(input_x.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
weight_mat = np.random.normal([2,1]).reshape([2,1])
print(weight_mat.shape)
print(weight_mat)
```

```python
def calc_sum_error(weight_mat,input_x,y):
    output = np.matmul(input_x,weight_mat)
#     print(output[:10,:])
    error = (output - y)**2/input_x.shape[0]
#     print(error[:10,:])
    return error.sum()
```

```python
error = calc_sum_error(weight_mat,input_x,y)
print(error)
```

```python
def calc_gred(weight_mat,input_x,y,lr):
#     print(input.shape)
    output = np.matmul(input_x,weight_mat)
    d_w = 2*np.multiply((output - y),x)/input_x.shape[0]
    d_b = 2*np.multiply((output - y),1)/input_x.shape[0]
#     print(d_w,d_b)
    return lr*d_w.sum(),lr*d_b.sum()
```

```python
d_w,d_b = calc_gred(weight_mat,input_x,y,lr)
print(d_w,d_b)
```

```python
import IPython
min_x = x.min()
max_x = x.max()
x_range = np.arange(min_x,max_x,0.5)
# m = plt.figure(figsize = (15,10))
for each in range(50):
    error = calc_sum_error(weight_mat,input_x,y)
    d_w,d_b = calc_gred(weight_mat,input_x,y,lr)
    gred = np.array([[d_w],[d_b]]).reshape([2,-1])
#     print(weight_mat)
#     print(gred)
    weight_mat = weight_mat - gred
#     print(weight_mat)
#     if each%100 == 0:
    print('Error: ',error)
    print('Gredient: ',d_w,d_b)
    print('Weight : ',weight_mat[0],weight_mat[1])
    y_range = [weight_mat[0]*m+weight_mat[1] for m in x_range]
    plt.plot(x_range,y_range,color = 'r')
    plt.scatter(x,y,color = 'b')
    plt.show()
    plt.clf()
```
