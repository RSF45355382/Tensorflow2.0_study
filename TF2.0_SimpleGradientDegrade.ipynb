```python
import tensorflow as tf 
import numpy as np 
```

```python
def himmelblau(x):
    return (x[0]**2+x[1]-11)**2+(x[0]+x[1]**2-7)**2
```

```python
x = np.arange(-6.,6,0.1)
y = np.arange(-6.,6,0.1)
print('x,y shape: ',x.shape,y.shape)
```

```python
X,Y = np.meshgrid(x,y)
print('X,Y shape: ',X.shape,Y.shape)
# 生成坐标平面
```

```python
Z = himmelblau([X,Y])
```

```python
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure('himmelblau')
ax = fig.gca(projection = '3d')
ax.plot_surface(X,Y,Z)
ax.view_init(60,-38)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
x = tf.constant([-4.,0.])
# 随机设定一个初始点
for step in range(200):
    with tf.GradientTape() as tape:
        tape.watch([x])
        y = himmelblau(x)
        # 计算函数
    grad = tape.gradient(y,[x])[0]
    # 计算相对于x的梯度
    x -= 0.01*grad
    
    if step%20 == 0:
        print('Step {},x = {}, y = {}'.format(step,x.numpy(),y.numpy()))
```
